# Simple classification

For this first simple network, we want to linearly classify an image into 3 shape categories (rectangle, disk, triangle). 
Thus, we will use a 1-layer neural network for this task.

We start by generating a train set.

```python
[X_train, Y_train] = generate_dataset_classification(300, 20)
Y_train = np_utils.to_categorical(Y_train, 3)

print(X_train.shape)
```

## SGD optimizer
We first try using SGD optimizer for this linear classification, with a single layer, and softmax activation.  
We choose the loss function to be categorical crossentropy. 

```python
model = Sequential()
model.add(Dense(3,  input_dim = 10000, activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer="SGD")
model.fit(X_train, Y_train, epochs=6, batch_size=32)
```

The model does not seem to converge on a few epochs with default parameters, therefore, we lower the learning rate by a factor 10.

```python
from keras import optimizers 
model = Sequential()
model.add(Dense(3,  input_dim = 10000, activation='softmax'))

sgd = optimizers.SGD(lr=0.001, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy',optimizer=sgd,metrics=['accuracy'])
model.fit(X_train, Y_train, epochs=30, batch_size=32)
```

## Adam optimizer
We now train the linear classifier with Adam optimizer. 


```python
modelAdam = Sequential()
modelAdam.add(Dense(3,  input_dim = 10000, activation='softmax'))
modelAdam.compile(loss='categorical_crossentropy',optimizer="Adam" ,metrics=['accuracy'])
modelAdam.fit(X_train, Y_train, epochs=25, batch_size=32)
```

Using optimizer Adam makes it faster to train the model, but the models perform equally well (if tuning the learning rate of SGD). Thus, for larger datasets, Adam would be preferable.

# Testing the models

We first check the models on 3 examples (one for each shape). 

```python
X_test1 = generate_a_rectangle()
X_test1 = X_test1.reshape(1, X_test1.shape[0])
X_test2 = generate_a_disk()
X_test2 = X_test2.reshape(1, X_test2.shape[0])
X_test3 = generate_a_triangle()[0]
X_test3 = X_test3.reshape(1, X_test3.shape[0])
print("Model with SGD optimizer \n =============================")
print(model.predict(X_test1))
print(model.predict(X_test2))
print(model.predict(X_test3))
print("Model with Adam optimizer \n =============================")
print(modelAdam.predict(X_test1))
print(modelAdam.predict(X_test2))
print(modelAdam.predict(X_test3))
```

```python
[X_test, Y_test] = generate_test_set_classification(300)
```

```python
print(modelAdam.evaluate(X_test, modelAdam.predict(X_test)))
print(model.evaluate(X_test, model.predict(X_test)))
```

Both models reach 100% accuracy.

# A More Difficult Classification Problem

In order to get better results, the size of the train set was increased to 1000 examples.

```python
[X_train1, Y_train1] = generate_dataset_classification(1000, 20, True)
Y_train1 = np_utils.to_categorical(Y_train1)
X_train1 = X_train1.reshape(1000,100,100,1)
```

```python
[X_test1, Y_test1] = generate_test_set_classification(500)
X_test1 = X_test1.reshape(500,100,100,1)
```

We build a more complex neural network, adding one convolutional layer with 64 5x5 filters, and a dense layer to the output layer. 

```python
model1 = Sequential()
model1.add(Conv2D(64, (5,5), input_shape = (100,100,1), activation='relu'))
print(model1.output_shape)
model1.add(MaxPooling2D(pool_size = (4,4)))
model1.add(Dropout(0.5))
model1.add(Flatten())
model1.add(Dense(124, activation='relu'))
model1.add(Dense(3, activation='softmax'))

model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
print(model1.summary())
```

```python
model1.fit(X_train1, Y_train1, epochs=16, batch_size=32)
```

```python
model1.evaluate(X_test1, Y_test1)
```

Accuracy on the test set is fairly good, 90%, this might be that we overfit the training test, one way to improve the result could be tu use early stopping. 

# Regression problem

The task now is to predict the image locations of the vertices of a triangle, given an image of this
triangle.

```python
[X_train2, Y_train2] = generate_dataset_regression(300, 20)
[X_test2, Y_test2] = generate_test_set_regression()
```

```python
visualize_prediction(X_train2[0], Y_train2[0])
```

The neural network is made of 3 convolutional layers and 2 dense layers, with activation function reLU. 
The loss function is mean absolute error, and optimizer is Adam. 


```python
from keras.layers import BatchNormalization
model2 = Sequential()
model2.add(Conv2D(64, (5,5), input_shape = (100,100,1), activation='relu'))
model2.add(MaxPooling2D(pool_size = (4,4)))
model2.add(Dropout(0.5))
model2.add(Conv2D(128, (5, 5), activation='relu'))
model2.add(Dropout(0.5))
model2.add(Conv2D(128, (5, 5), activation='relu'))
model2.add(Flatten())
model2.add(Dense(64, activation='relu'))
model2.add(Dropout(0.1))
#model2.add(BatchNormalization())
model2.add(Dense(6))
print(model2.output_shape)
model2.compile(loss='mean_absolute_error', optimizer='adam')
print(model2.summary())
```

We sort the vertices by x and y coordinates. 

```python
def sort_vertices(dataset):
    res = np.zeros((300, 6))
    for i in range(0, 300):
        temp1 = [dataset[i,2*m] for m in range(0, 3)]
        j = np.argmax(temp1)
        k = np.argmin(temp1)
        l = 3 - j - k
        temp2 = [dataset[i,2*j], dataset[i,2*j+1], 
                dataset[i,2*k], dataset[i,2*k+1], 
                dataset[i,2*l], dataset[i,2*l+1]]
        if (dataset[i,2*j] == dataset[i,2*k] and dataset[i,2*j+1]<dataset[i,2*k+1]):
            temp2 = [dataset[i,2*k], dataset[i,2*k+1],
                     dataset[i,2*j], dataset[i,2*j+1],  
                    dataset[i,2*l], dataset[i,2*l+1]]
        elif (dataset[i,2*l] == dataset[i,2*k] and dataset[i,2*k+1]<dataset[i,2*l+1]):
            temp2 = [dataset[i,2*j], dataset[i,2*j+1],
                dataset[i,2*l], dataset[i,2*l+1],
                dataset[i,2*k], dataset[i,2*k+1]]
        res[i,:] = temp2
    return res
```

```python
Y_train3 = sort_vertices(Y_train2)
Y_test3 = sort_vertices(Y_test2)
```

```python
X_train2 = X_train2.reshape(300,100,100,1)
```

```python
model2.fit(X_train2, Y_train3, epochs=40, batch_size=32)
```

```python
X_test2 = X_test2.reshape(300,100,100,1)
```

```python
model2.evaluate(X_test2, Y_test3)
```

```python
visualize_prediction(X_test2[0],model2.predict(X_test2)[0])
```

```python
visualize_prediction(X_test2[1],model2.predict(X_test2)[1])
```

```python
visualize_prediction(X_test2[4],model2.predict(X_test2)[4])
```

With this model, the mean absolute error on the test set is 0.1 (mean squared error is about 0.02), the triangles fall in the regions predicted.

Bonus question

We build an hourglass network to denoise an image. 
In each test set example, we add a random noise (random integer chosen between 0 and 255) and return both noisy figure and clean figure (formulas are those in the generating figures functions in the mp1.py file)

I chose to use convolutional layers, and my final architecture has 7 layers. 

```python
from keras.layers import UpSampling2D

modelHour = Sequential()
modelHour.add(Conv2D(16, (3, 3), activation='relu', input_shape = (100,100,1)))
modelHour.add( MaxPooling2D(pool_size = (2, 2)))
modelHour.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
modelHour.add(MaxPooling2D((2, 2), padding='same'))
modelHour.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
modelHour.add(MaxPooling2D((2, 2), padding='same'))
modelHour.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
modelHour.add(UpSampling2D((2, 2)))
modelHour.add(Conv2D(8, (3, 3), activation='relu', padding='same'))
modelHour.add(UpSampling2D((2, 2)))
modelHour.add(Conv2D(16, (3, 3), activation='relu'))
modelHour.add(UpSampling2D((2, 2)))
modelHour.add(Conv2D(1, (3, 3), activation='sigmoid', padding='same'))

modelHour.compile(loss='binary_crossentropy', optimizer='adadelta')
```

```python
x_train_noise, x_train_clean = generate_dataset_classification_noise(300)
```

```python
x_train_noise = np.reshape(x_train_noise, (len(x_train_noise), 100, 100, 1))  
x_train_clean = np.reshape(x_train_clean, (len(x_train_clean), 100, 100, 1))
```

```python
x_test_noise, x_test_clean = generate_dataset_classification_noise(300)
```

```python
x_test_noise = np.reshape(x_test_noise, (len(x_test_noise), 100, 100, 1))  
x_test_clean = np.reshape(x_test_clean, (len(x_test_clean), 100, 100, 1))
```

```python
modelHour.fit(x_train_noise, x_train_clean,
                epochs=30,
                batch_size=32)
```

```python
modelHour.evaluate(x_test_noise, x_test_clean)
```

```python
plt.figure(figsize=(50, 50))

ax = plt.subplot(1, 2, 1)
plt.imshow(x_test_noise[1].reshape(100, 100))
ax = plt.subplot(1, 2, 2)
plt.imshow(modelHour.predict(x_test_noise)[1].reshape(100, 100))
plt.gray()
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.show()
```

We are able to denoise the image, we lose a little bit of accuracy in the corners of the rectangle,but overall, the result is pretty satisfactory.
